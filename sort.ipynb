```python
def three_sort(nums):
    i = 0
    j = len(nums)-1
    while i<j:
        if nums[i] % 3 == 0:
            nums[i],nums[j] = nums[j], nums[i]
            i += 1
        elif nums[j] % 3 == 0:
            j -= 1
        else:
            i += 1 
    return nums   
```

```python
three_sort([3,9,12,7,5,2])
```
